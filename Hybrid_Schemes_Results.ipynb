```python
results_dir = 'results'
problems = ['adlittle', 'afiro', 'agg2','agg3','agg', 'bandm',
 'beaconfd', 'blend', 'bnl1', 'boeing1', 'boeing2', 'bore3d', 'brandy', 'capri',
 'cycle', 'degen2', 'degen3', 'e226', 'etamacro', 'fffff800', 'finnis',
 'fit1d', 'fit1p', 'forplan', 'ganges', 'gfrd-pnc', 'grow15', 'grow22', 'grow7',
 'israel', 'kb2', 'ken-07', 'lotfi', 'maros', 'pds-02', 'perold',
 'pilotnov', 'qap08', 'recipe', 'sc105', 'sc205', 'sc50a',
 'sc50b', 'scagr25', 'scagr7', 'scfxm1', 'scfxm2', 'scfxm3', 'scorpion', 'scrs8',
 'scsd1', 'scsd6', 'scsd8', 'sctap1', 'sctap2', 'sctap3', 'seba', 'share1b',
 'share2b', 'shell', 'ship04l', 'ship04s', 'ship08l', 'ship08s', 'ship12l', 'ship12s',
 'sierra', 'stair', 'standata', 'standgub', 'standmps', 'stocfor1', 'stocfor2', 'tuff',
 'vtp_base', 'wood1p', 'cre-c', 'modszk1']
 # 'pilot4',

avgs = {'lp_time': [], 'old_sd_time': [], 'old_avg_sd': [],
        'alt_sd_time': [], 'avg_alt_sd': [],
        'alt_hybrid_up_act_sd_time': [], 'alt_hybrid_up_act_avg_sd': [], 
        'alt_hybrid_init_act_sd_time': [],'alt_hybrid_init_act_avg_sd': [],
        'reg_hybrid_up_act_sd_time': [], 'reg_hybrid_up_act_avg_sd': [], 
        'reg_hybrid_init_act_sd_time': [],'reg_hybrid_init_act_avg_sd': [],
        'lp_iters': [], 'old_sd_iters': [],
        'alt_sd_iters': [],
        'alt_hybrid_up_act_sd_iters': [], 'alt_hybrid_init_act_sd_iters': [],
       'reg_hybrid_up_act_sd_iters': [], 'reg_hybrid_init_act_sd_iters': []}
from statistics import median

figure_dir = 'figures'
if not os.path.exists(figure_dir): os.mkdir(figure_dir)
for problem in problems:
    lp_fn = os.path.join(results_dir, problem + '_lp.p')
    old_sd_fn = os.path.join(results_dir, problem + '_old_sd.p')
    if not os.path.exists(old_sd_fn): continue
    alt_sd_fn = os.path.join(results_dir, problem + '_alt_sd.p')
    if not os.path.exists(alt_sd_fn): continue
    alt_hybrid_up_act_sd_fn = os.path.join(results_dir, problem + '_alt_hybrid_up_act_sd.p')
    if not os.path.exists(alt_hybrid_up_act_sd_fn): continue
    alt_hybrid_init_act_sd_fn = os.path.join(results_dir, problem + '_alt_hybrid_init_act_sd.p')
    if not os.path.exists(alt_hybrid_init_act_sd_fn): continue
    reg_hybrid_up_act_sd_fn = os.path.join(results_dir, problem + '_reg_hybrid_up_act_sd.p')
    if not os.path.exists(reg_hybrid_up_act_sd_fn): continue
    reg_hybrid_init_act_sd_fn = os.path.join(results_dir, problem + '_reg_hybrid_init_act_sd.p')
    if not os.path.exists(reg_hyrbid_init_act_sd_fn): continue
    
    
    with open(lp_fn, 'rb') as f:
        lp_results = pickle.load(f)
    with open(old_sd_fn, 'rb') as f:
        old_sd_results = pickle.load(f)
    with open(alt_sd_fn, 'rb') as f:
        alt_sd_results = pickle.load(f)
    with open(alt_hybrid_up_act_sd_fn, 'rb') as f:
        alt_hybrid_up_act_sd_result = pickle.load(f)
    with open(alt_hybrid_init_act_sd_fn, 'rb') as f:
        alt_hybrid_init_act_sd_result = pickle.load(f)
    with open(reg_hybrid_up_act_sd_fn, 'rb') as f:
        reg_hybrid_up_act_sd_result = pickle.load(f)
    with open(reg_hybrid_init_act_sd_fn, 'rb') as f:
        simp_reg_hybrid_act_sd_result = pickle.load(f)
        
    try: 
        avgs['old_avg_sd'].append(avg(old_sd_results['solve_times'][1:]))
        avgs['alt_avg_sd'].append(avg(alt_sd_results['solve_times'][1:]))
        avgs['alt_hybrid_up_act_avg_sd'].append(avg(alt_hybrid_up_act_sd_result['solve_times'][1:]))
        avgs['alt_hybrid_init_act_avg_sd'].append(avg(alt_hybrid_init_act_sd_result['solve_times'][1:]))
        avgs['reg_hybrid_up_act_avg_sd'].append(avg(reg_hybrid_up_act_sd_result['solve_times'][1:]))
        avgs['reg_hybrid_init_act_avg_sd'].append(avg(reg_hybrid_init_act_sd_result['solve_times'][1:]))
        avgs['lp_time'].append(lp_results['solve_time_total'])
        avgs['old_sd_time'].append(old_sd_results['solve_time_total'])
        avgs['alt_sd_time'].append(alt_sd_results['solve_time_total'])
        avgs['alt_hybrid_up_act_sd_time'].append(alt_hybrid_up_act_sd_result['solve_time_total'])
        avgs['alt_hybrid_init_act_sd_time'].append(alt_hybrid_init_act_sd_result['solve_time_total'])
        avgs['reg_hybrid_up_act_sd_time'].append(reg_hybrid_up_act_sd_result['solve_time_total'])
        avgs['reg_hybrid_init_act_sd_time'].append(reg_hybrid_init_act_sd_result['solve_time_total'])
        avgs['lp_iters'].append(lp_results['n_iters'])
        avgs['old_sd_iters'].append(old_sd_results['n_iters'])
        avgs['alt_sd_iters'].append(alt_sd_results['n_iters'])
        avgs['alt_hybrid_up_act_sd_iters'].append(alt_hybrid_up_act_sd_result['n_iters'])
        avgs['alt_hybrid_init_act_sd_iters'].append(alt_hybrid_init_act_sd_result['n_iters'])
        avgs['reg_hybrid_up_act_sd_iters'].append(reg_hybrid_up_act_sd_result['n_iters'])
        avgs['reg_hybrid_init_act_sd_iters'].append(reg_hybrid_init_act_sd_result['n_iters'])
    except KeyError:
        print('skipping {}'.format(problem))
        continue
        
for metric in avgs:
    print(metric)
    print('Average {}: {}'.format(metric, avg(avgs[metric])))
    print('Median {}: {}'.format(metric, median(avgs[metric])))
```
